```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# reading dataset
df = pd.read_excel('date-after-preprocessing.xlsx')
df = df.iloc[:,1:]
df = df.drop(['PAC','PAC Dev','project_name'],axis=1)
df = df.replace(0,np.nan)
df.head(5)
```

```python
df.describe()
```

```python
# pairplot
sns.pairplot(df)
plt.subplots_adjust(left=0.05,right=0.95,bottom=0.05,top=0.95)
plt.savefig('pairplot.png')
```

```python
# 3d plot
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(df['Duration'],df['Total Cost'],df['Actual Cost'])
plt.savefig('3d.png')
plt.show()
```
