```python
HLWBS = {}
for file in msp_files:
    df = pd.read_excel('C:/Users/Reza/Desktop/Bachelor Project/Dataset/Data set/'+file, sheet_name='Baseline Schedule',header=None)
    df = df.iloc[1:,:]
    headers = df.iloc[0]
    df =df[1:]
    df.columns = headers
    for index,row in df.iterrows():
        if pd.isnull(row['Predecessors']) and pd.isnull(row['Successors']):
            HLWBS[row['Name']] = row['ID']
            print(row['Name']+ " "+ str(row['ID']))
            
print(HLWBS)
```

```python
with open('hlwbs.csv', 'w',newline='') as hlbws_file:
    writer = csv.writer(hlbws_file,delimiter=',')
    for key,value in HLWBS.items():
        writer.writerow((key,value))
```
